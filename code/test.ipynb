根据相关性可以看出比较重要的两个负相关特征：净利润平均值和纳税总额平均值。<br>
下面绘制这两个特征的箱线图查看其分布情况。

```python
pos_data = train_data[train_data.flag == 1]
neg_data = train_data[train_data.flag == -1]
```

```python
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
sns.set()
%config InlineBackend.figure_format = 'retina' 
```

```python
fig, axes = plt.subplots(1, 2, figsize=(15, 5))

axes[0].boxplot([neg_data["净利润平均值"], pos_data["净利润平均值"]], widths=0.7)
axes[0].set_xlabel('Flag')
axes[0].set_xticks([1, 2])
axes[0].set_xticklabels(['-1', '1']);

axes[1].boxplot([neg_data["纳税总额平均值"], pos_data["纳税总额平均值"]], widths=0.7)
axes[1].set_xlabel('Flag')
axes[1].set_xticks([1, 2])
axes[1].set_xticklabels(['-1', '1']);
```

```python
pos_data_val = verify_data[verify_data.flag == 1]
neg_data_val = verify_data[verify_data.flag == -1]
```

```python
fig, axes = plt.subplots(1, 2, figsize=(15, 5))

axes[0].boxplot([neg_data_val["净利润平均值"], pos_data_val["净利润平均值"]], widths=0.7)
axes[0].set_xlabel('Flag')
axes[0].set_xticks([1, 2])
axes[0].set_xticklabels(['-1', '1']);

axes[1].boxplot([neg_data_val["纳税总额平均值"], pos_data_val["纳税总额平均值"]], widths=0.7)
axes[1].set_xlabel('Flag')
axes[1].set_xticks([1, 2])
axes[1].set_xticklabels(['-1', '1']);
```

从图中可以看到，正负类在净利润平均值这个特征的区别不大，但在纳税总额均值这个特征上区别明显。<br>
下面用线性分类器分别对相关性最高的两个特征、相关性最高的十个特征和全部特征进行训练。

```python
mean_data = train_data[["净利润平均值", "纳税总额平均值", "flag"]]
```

```python
mean_data
```

```python
from sklearn.linear_model import SGDClassifier
```

```python
clf = SGDClassifier(max_iter=5000)
```

```python
X = mean_data.drop(["flag"], axis=1)
y = mean_data.flag
```

```python
clf.fit(X, y)
```

```python
from sklearn.metrics import accuracy_score, recall_score, f1_score, confusion_matrix
```

```python
y_pred = clf.predict(X)
```

```python
def plot(y_true, y_pred):
    sns.heatmap(confusion_matrix(y_true, y_pred), square=True, annot=True, fmt='g', cbar=False, annot_kws={'size': 30})
    plt.xlabel("Predicted")
    plt.ylabel("True")
    P = accuracy_score(y_true, y_pred) * 100
    R = recall_score(y_true, y_pred) * 100
    F = f1_score(y_true, y_pred) * 100
    plt.title("P: {:.2f}, R: {:.2f}, F: {:.2f}".format(P, R, F))
```

```python
plot(y, y_pred)
```

```python
p = train_data[train_data.flag == 1].index.tolist()
u = train_data[train_data.flag == -1].index.tolist()
```

```python
row_rand_array = np.arange(len(p))
np.random.shuffle(row_rand_array)
n = int(len(p)*0.15)
s = row_rand_array[:n] 
```

```python
y_mod = y.copy()
```

```python
y_mod[s] = -1
```

```python
pu_clf = SGDClassifier(loss='log')
```

```python
pu_clf.fit(X_all, y_mod)
```

```python
pred_ys = pu_clf.predict_proba(X_all.iloc[s, :])
```

```python
for i in range(0, 100):
    tr = i/100
    acc = len(np.argwhere(pred_ys[:, 1] > tr)) / pred_ys.shape[0]
    if acc < 0.975:
        tr -= 0.01
        break
```

```python
tr
```

```python
pred_yu = pu_clf.predict_proba(X_all.iloc[u, :])
```

```python
rn = np.argwhere(pred_yu[:, 1] < tr).squeeze()
```

```python
rn.shape
```

```python
X_new = pd.concat([X_all.iloc[p, :], X_all.iloc[rn, :]])
```
